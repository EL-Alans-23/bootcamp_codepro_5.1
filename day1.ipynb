```python
#Imagina que estás repartiendo caramelos: solicita al usuario ingresar el total de caramelos y la cantidad de niños, luego calcula cuántos caramelos recibe cada niño usando división entera y cuántos sobran usando módulo, e imprime ambos resultados. 
cantidad_mitai = int(input('ingrese la cantidad de mitai'))
cantidad_caramelos = int(input('ingrese la cantidad de caramelos'))

division_caramelos = cantidad_caramelos // cantidad_mitai
caramelos_restantes = cantidad_caramelos % cantidad_mitai

print(f'la cantidad de caramelo por mitai es {division_caramelos}')
print(f'la cantidad de caramelo resante es {caramelos_restantes}')
```

```python
#funciones
def suma_altura():
    nombre = input('ingrese su nombre')
    altura= int(input("ingresa su altura en cm"))
    nombre_companero = input('ingrese el nombre de su companero')
    altura_companero = int(input('ingresa la altura de su companero en cm'))
    suma_altura= altura_companero + altura
    resta_altura = altura - altura_companero
    print(f'la suma de alturas de {nombre} y {nombre_companero} = {suma_altura}cm y la diferencia de alturas entre {nombre} y {nombre_companero} = {resta_altura}cm')

suma_altura()
suma_altura()


```
