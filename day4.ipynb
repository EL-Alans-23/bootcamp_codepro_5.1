```python
biografia = {}

def biografia_persona():
    biografia ['nombre'] = input('ingrese su nombre')
    print(biografia)
    biografia ['sexo'] = input('ingrese su sexo')
    print(biografia)
    biografia ['edad'] = int(input('ingrese su edad'))
    print(biografia)
    biografia ['telefono'] = int(input('ingrese tu telefono'))
    print(biografia)
    biografia ['mail'] = input('ingrese su mail')
    print(biografia)

for i in range(5):
    biografia_persona()
```

```python
datos_tarjeta = {}

while True:
    print('--menu--')
    print('1.Crear datos nuevos')
    print('2.Modificar datos existentes')
    print('3.Eliminar datos existentes')
    print('4.Ver los Datos')
    print('5.Salir')
    opcion = int(input('ingrese una opcion'))

    if opcion == 1:
        datos_tarjeta['Nombre'] = input('Ingrese su nombre')
        datos_tarjeta['Apellido'] = input('Ingrese su apellido')
        datos_tarjeta['Edad'] = input('Ingrese su edad')
    elif opcion == 2:
        print('--menu--')
        print('Que desea modificar')
        print('1.Nombre')
        print('2.Apellido')
        print('3.Edad')
        opcion1 = int(input('ingrese una opcion'))
        if opcion1 == 1:
            datos_tarjeta['Nombre'] = input('Ingrese su nombre')
        elif opcion1 == 2:
            datos_tarjeta['Apellido'] = input('Ingrese su apellido')
        elif opcion1 == 3:
            datos_tarjeta['Edad'] = input('Ingrese su edad')
        else:
            print('Ingrese una opcion valida')
    elif opcion == 3:

        print('--menu--')
        print('Que desea eliminar')
        print('1.Nombre')
        print('2.Apellido')
        print('3.Edad')
        opcion2 = int(input('ingrese una opcion'))
        if opcion2 == 1:
            del datos_tarjeta['Nombre']
        elif opcion2 == 2:
            del datos_tarjeta['Apellido']
        elif opcion2 == 3:
            del datos_tarjeta['Edad']
        else:
            print('Ingrese una opcion valida')
    elif opcion == 4:
        print(datos_tarjeta)
    elif opcion == 5:
        print('Chau')
        break

    else:
        print('ingrese una opcion valida')
```

```python
contactos = []

contacto = {}
def mostrar():
    print(f'Tus contactos son: {contactos}')

def crear():
    clave_numero = int(input('ingrese su numero'))
    clave_nombre = input('Igrese el nombre')
    contacto[clave_nombre] = clave_numero
    contactos.append(contacto)


def modificar():
    contacto[input('Igrese el nombre')] = int(input('ingrese su numero'))

while True:
    print('--menu--')
    print('1.Crear contacto')
    print('2.mostrar contactos')
    print('3.Modificar el contacto')
    print('4.Salir')
    opcion = int(input('ingrese una opcion'))
    if opcion == 1:
        crear()
    elif opcion == 2:
        mostrar()
    elif opcion == 3:
        modificar()
    elif opcion == 4:
        print('chau')
        break
    else:
        print('Ingrese una opcion valida')
```
