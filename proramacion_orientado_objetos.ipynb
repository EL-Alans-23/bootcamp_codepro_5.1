```python
class personaje():
    
    def __init__(self,nombre,ataque,vida):
        self.nombre = nombre
        self.ataque = ataque
        self.vida = vida
    
    def atributos (self):
        print(self.nombre )
        print("vida:",self.vida)
        print("ataque:",self.ataque)
    
    def morir (self,enemigo):
        if self.vida <= 0 or enemigo.vida <= 0:
            print('esta muerto')
        else:
            print('esta vivo')
    
    def atacar (self,enemigo,poder):
        enemigo.vida -= poder

class mago(personaje):
    def __init__(self, nombre, ataque, vida,inteligencia):
        super().__init__(nombre, ataque, vida)
        self.inteligencia = inteligencia
    
    def atributos(self):
        print(self.nombre )
        print("vida:",self.vida)
        print("ataque:",self.ataque)
        print('inteligencia:',self.inteligencia)

    def poder_aumentado(self):
        poder = self.ataque * self.inteligencia

class barbaro(personaje):
    def __init__(self, nombre, ataque, vida,espada):
        super().__init__(nombre, ataque, vida)
        self.espada = espada 

    def poder_aumentado(self):
        poder = self.ataque * self.espada
    
    def atributos(self):
        print(self.nombre )
        print("vida:",self.vida)
        print("ataque:",self.ataque)
        print('espada:',self.espada)

#class tablero(personaje):
 #   def turnos(jugador1,jugador2):
  #      intentos = 0
   #     while True:
    #        jugador1.atacar(jugador2)
     #       jugador2.atacar(jugador1)
      #      intentos = intentos + 1
       #     if jugador1.vida() <= 0 :
        #        jugador1.morir()
         #       break
          #  elif jugador2.vida() <= 0:
           #     jugador2.morir()
            #    break

jugador1 = mago('mago',3,100,5)
jugador2 = barbaro('barbaro',5,100,2)
jugador1.atributos()
print("")
jugador2.atributos()
print("")
jugador1.atacar(jugador2,jugador1.poder_aumentado())
```
